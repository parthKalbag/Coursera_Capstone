```python
df = pd.read_csv('../datasets/US_Accidents_June20.csv')
df.head(5)
```

<h2>Features of the dataset</h2>

<h4>Traffic Attributes (12):</h4>
<ol>
    <li>ID: This is a unique identifier of the accident record</li>
    <li>Source: Indicates source of the accident report (i.e. the API which reported the accident.)</li>
    <li>TMC: A traffic accident may have a Traffic Message Channel (TMC) code which provides more detailed description of the event</li>
    <li>Severity: Shows the severity of the accident, a number between 1 and 4, where 1 indicates the least impact on traffic (i.e., short delay as a result of the accident) and 4 indicates a significant impact on traffic (i.e., long delay).
    <li>Start_Time: Shows start time of the accident in local time zone.
    <li>End_Time: Shows end time of the accident in local time zone.
    <li>Start_Lat: Shows latitude in GPS coordinate of the start point.
    <li>Start_Lng: Shows longitude in GPS coordinate of the start point.
    <li>End_Lat: Shows latitude in GPS coordinate of the end point.
    <li>End_Lng: Shows longitude in GPS coordinate of the end point.
    <li>Distance(mi): The length of the road extent affected by the accident.
    <li>Description: Shows natural language description of the accident
</ol>

<h4>Address Attributes (9):</h4>
<ol>
    <li>Number: Shows the street number in address field.
    <li>Street: Shows the street name in address field.
    <li>Side: Shows the relative side of the street (Right/Left) in address field.
    <li>City: Shows the city in address field.
    <li>County: Shows the county in address field.
    <li>State: Shows the state in address field.
    <li>Zipcode: Shows the zipcode in address field.
    <li>Country: Shows the country in address field.
    <li>Timezone: Shows timezone based on the location of the accident (eastern, central, etc.).
</ol>

<h4>Weather Attributes (11):</h4>
<ol>
    <li>Airport_Code: Denotes an airport-based weather station which is the closest one to location of the accident.
    <li>Weather_Timestamp: Shows the time-stamp of weather observation record (in local time).
    <li>Temperature(F): Shows the temperature (in Fahrenheit).
    <li>Wind_Chill(F): Shows the wind chill (in Fahrenheit).
    <li>Humidity(%): Shows the humidity (in percentage).
    <li>Pressure(in): Shows the air pressure (in inches).
    <li>Visibility(mi): Shows visibility (in miles).
    <li>Wind_Direction: Shows wind direction.
    <li>Wind_Speed(mph): Shows wind speed (in miles per hour).
    <li>Precipitation(in): Shows precipitation amount in inches, if there is any.
    <li>Weather_Condition: Shows the weather condition (rain, snow, thunderstorm, fog, etc.).
</ol>

<h4>POI Attributes (13):</h4>
<ol>
    <li>Amenity: A Point-Of-Interest (POI) annotation which indicates presence of amenity in a nearby location.
    <li>Bump: A POI annotation which indicates presence of speed bump or hump in a nearby location.
    <li>Crossing: A POI annotation which indicates presence of crossing in a nearby location.
    <li>Give_Way: A POI annotation which indicates presence of give_way sign in a nearby location.
    <li>Junction: A POI annotation which indicates presence of junction in a nearby location.
    <li>No_Exit: A POI annotation which indicates presence of no_exit sign in a nearby location.
    <li>Railway: A POI annotation which indicates presence of railway in a nearby location.
    <li>Roundabout: A POI annotation which indicates presence of roundabout in a nearby location.
    <li>Station: A POI annotation which indicates presence of station (bus, train, etc.) in a nearby location.
    <li>Stop: A POI annotation which indicates presence of stop sign in a nearby location.
    <li>Traffic_Calming: A POI annotation which indicates presence of traffic_calming means in a nearby location.
    <li>Traffic_Signal: A POI annotation which indicates presence of traffic_signal in a nearby location.
    <li>Turning_Loop: A POI annotation which indicates presence of turning_loop in a nearby location.
</ol>
<h4>Period-of-Day (4):</h4>
<ol>
    <li>Sunrise_Sunset: Shows the period of day (i.e. day or night) based on sunrise/sunset.
    <li>Civil_Twilight: Shows the period of day (i.e. day or night) based on civil twilight.
    <li>Nautical_Twilight: Shows the period of day (i.e. day or night) based on nautical twilight.
    <li>Astronomical_Twilight: Shows the period of day (i.e. day or night) based on astronomical twilight.
</ol>

<p>Note that the data has come from two different sources MapQuest and Bing</p>
<p>Since severity is what we really care about in this project, I think it is crucial to figure out the difference.</p>

```python
df_source = df.groupby(['Severity','Source']).size().reset_index().pivot(\
    columns='Severity', index='Source', values=0)
df_source.plot(kind='bar', stacked=True, title='Severity Count by Sources')
df_source
```

The stacked bar chart shows that two data providers reported totally different proportions of accidents of each level. MapQuest reported so rare accidents with severity level 4 which can not even be seen in the plot, whereas Bing reported almost the same number of level 4 accidents as level 2. Meanwhile, MapQuest reported much more level 3 accidents than Bing in terms of proportion. These differences may be due to the different kinds of accidents they tend to collect or the different definitions of severity level, or the combination of them. If the latter is the case, I don't think we can use the data from both of them at the same time. To check it out, we can examine the distribution of accidents with different severity levels across two main measures, Impacted Distance and Duration

```python
# fixing date-time
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = pd.to_datetime(df['End_Time'])
df['Weather_Timestamp'] = pd.to_datetime(df['Weather_Timestamp'])

# Calculating duration and adding it to the dataframe as a column
df['Duration'] = df.End_Time - df.Start_Time 
df['Duration'] = df['Duration'].apply(lambda x:round(x.total_seconds() / 60) )

fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
sns.boxplot(x="Severity", y="Duration",
            data=df.loc[(df['Source']=="MapQuest") & (df['Duration']<400),], palette="Set2", ax=axs[0])
axs[0].set_title('MapQuest')
fig.suptitle('Accidents Duration by Severity', fontsize=16)
sns.boxplot(x="Severity", y="Duration",
            data=df.loc[(df['Source']=="Bing") & (df['Duration']<400),], palette="Set2", ax=axs[1])
axs[1].set_title('Bing')
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
sns.boxplot(x="Severity", y="Distance(mi)", 
            data=df.loc[(df['Source']=="MapQuest") & (df['Distance(mi)']<5),], palette="Set2", ax=axs[0])
axs[0].set_title('MapQuest')
fig.suptitle('Impacted Distance by Severity', fontsize=16)
sns.boxplot(x="Severity", y="Distance(mi)",
            data=df.loc[(df['Source']=="Bing") & (df['Distance(mi)']<5),], palette="Set2", ax=axs[1])
axs[1].set_title('Bing')
plt.show()
```

Two differences are obvious in the above plots. The first is that the overall duration and impacted distance of accidents reported by Bing are much longer than those by MapQuest. Second, same severity level holds different meanings for MapQuest and Bing. MapQuest seems to have a clear and strict threshold for severity level 4, cases of which nevertheless only account for a tiny part of the whole dataset. Bing, on the other hand, doesn't seem to have a clear-cut threshold, especially regards duration, but the data is more balanced.

It is hard to choose one and we definitely can't use both. I decided to select MapQuest because serious accidents are we really care about and the sparse data of such accidents is the reality we have to confront.

Finally, drop data reported from Bing and 'Source' column.

```python
df = df.loc[df['Source']=="MapQuest",]
df = df.drop(['Source'], axis=1)
```

<h3>Removing Useless features</h3>
<p>Features 'ID' doesn't provide any useful information about accidents themselves. 'TMC', 'Distance(mi)', 'End_Time' (we have start time), 'Duration', 'End_Lat', 'Country', and 'End_Lng'(we have start location) can be collected only after the accident has already happened and hence cannot be predictors for serious accident prediction. For 'Description', the POI features have already been extracted from it by dataset creators. Let's get rid of these features first</p>

```python
df = df.drop(['ID', 'TMC', 'End_Time', 'Duration', 'End_Lat', 'Country', 'End_Lng', 'Description'], axis=1)
```

```python
df.shape
```

<h3>Cleaning of some categorical features</h3>

```python
df['Wind_Direction'].value_counts()
```

Now in the above dataframe the (East and E, ESE, ENE), (South and S,SSW, SSE), (North and N, NNW, NNE), (West and WSW, WNW, W) are basically the same thing so we must convert all of them to a standard variable

```python
df.loc[df['Wind_Direction']=='Calm','Wind_Direction'] = 'CALM'
df.loc[(df['Wind_Direction']=='West')|(df['Wind_Direction']=='WSW')|(df['Wind_Direction']=='WNW'),'Wind_Direction'] = 'W'
df.loc[(df['Wind_Direction']=='South')|(df['Wind_Direction']=='SSW')|(df['Wind_Direction']=='SSE'),'Wind_Direction'] = 'S'
df.loc[(df['Wind_Direction']=='North')|(df['Wind_Direction']=='NNW')|(df['Wind_Direction']=='NNE'),'Wind_Direction'] = 'N'
df.loc[(df['Wind_Direction']=='East')|(df['Wind_Direction']=='ESE')|(df['Wind_Direction']=='ENE'),'Wind_Direction'] = 'E'
df.loc[df['Wind_Direction']=='Variable','Wind_Direction'] = 'VAR'

df['Wind_Direction'].value_counts()
```

```python
df['Weather_Condition'].unique().tolist()
```

Let's collect some common weather conditions and drop the column

```python
df['Clear'] = np.where(df['Weather_Condition'].str.contains('Clear', case=False, na = False), 1, 0)
df['Cloud'] = np.where(df['Weather_Condition'].str.contains('Cloud|Overcast', case=False, na = False), 1, 0)
df['Rain'] = np.where(df['Weather_Condition'].str.contains('Rain|storm', case=False, na = False), 1, 0)
df['Heavy_Rain'] = np.where(df['Weather_Condition'].str.contains('Heavy Rain|Rain Shower|Heavy T-Storm|Heavy Thunderstorms', case=False, na = False), 1, 0)
df['Snow'] = np.where(df['Weather_Condition'].str.contains('Snow|Sleet|Ice', case=False, na = False), 1, 0)
df['Heavy_Snow'] = np.where(df['Weather_Condition'].str.contains('Heavy Snow|Heavy Sleet|Heavy Ice Pellets|Snow Showers|Squalls', case=False, na = False), 1, 0)
df['Fog'] = np.where(df['Weather_Condition'].str.contains('Fog', case=False, na = False), 1, 0)

# Assign NA to created weather features where 'Weather_Condition' is null.
weather = ['Clear','Cloud','Rain','Heavy_Rain','Snow','Heavy_Snow','Fog']
for i in weather:
  df.loc[df['Weather_Condition'].isnull(),i] = df.loc[df['Weather_Condition'].isnull(),'Weather_Condition']

df.loc[:,['Weather_Condition'] + weather]

df = df.drop(['Weather_Condition'], axis=1)
```

```python
# Fixing datetime format
# As the Weather_Timestamp is basically the same as that of Start_Time and we only need year, month, day, hour, etc. from them we can drop the column

df = df.drop(["Weather_Timestamp"], axis=1)

df['Year'] = df['Start_Time'].dt.year

nmonth = df['Start_Time'].dt.month
df['Month'] = nmonth

df['Weekday']= df['Start_Time'].dt.weekday

days_each_month = np.cumsum(np.array([0,31,28,31,30,31,30,31,31,30,31,30,31]))
nday = [days_each_month[arg-1] for arg in nmonth.values]
nday = nday + df["Start_Time"].dt.day.values
df['Day'] = nday

df['Hour'] = df['Start_Time'].dt.hour

df['Minute']=df['Hour']*60.0+df["Start_Time"].dt.minute

df.loc[:4,['Start_Time', 'Year', 'Month', 'Weekday', 'Day', 'Hour', 'Minute']]
```

<h3>Handling Missing values</h3>

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Now as far as the Number, Wind_Chill(F), and Precipitation are concerned more than 70 % of the data is missing and we can't just dropna as we would lose that data.  'Number' and 'Wind_Chill(F)' will be dropped because they are not highly related to severity according to previous research, whereas 'Precipitation(in)' could be a useful predictor and hence can be handled by separating feature.

```python
df = df.drop(['Number','Wind_Chill(F)'], axis=1)
```

```python
# Replacing the values of Precipitation with it's median
df['Precipitation_NA'] = 0
df.loc[df['Precipitation(in)'].isnull(),'Precipitation_NA'] = 1
df['Precipitation(in)'] = df['Precipitation(in)'].fillna(df['Precipitation(in)'].median())
df.loc[:5,['Precipitation(in)','Precipitation_NA']]
```

And as for the remaining values we can't substitute anything regarding it so we can drop the rows containing the values

```python
df = df.dropna(subset=['City','Zipcode','Airport_Code',
                       'Sunrise_Sunset','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight'])
```

```python
df.isnull().sum()
```

```python
Weather_data=['Temperature(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)']
print("The number of remaining missing values: ")
for i in Weather_data:
  df[i] = df.groupby(['Airport_Code','Month'])[i].apply(lambda x: x.fillna(x.median()))
  print( i + " : " + df[i].isnull().sum().astype(str))
```

```python
# There are some missing values in weather data but they are very less so we can drop those
df.dropna(subset=Weather_data, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Now for weather features mode will better rather than median

from collections import Counter
weather_cat = ['Wind_Direction'] + weather
print("Count of missing values that will be dropped: ")
for i in weather_cat:
  df[i] = df.groupby(['Airport_Code','Month'])[i].apply(lambda x: x.fillna(Counter(x).most_common()[0][0]) if all(x.isnull())==False else x)
  print(i + " : " + df[i].isnull().sum().astype(str))

# drop na
df = df.dropna(subset=weather_cat)
```

```python
df.isnull().sum()
```

<h3>Modeling </h3>
<p> Hence from the above data analysis the variables which have most impact on Accident Severity are: </p>
<ul>
    <li>Distance(mi)</li>
    <li>Start_Time and Start_Lat as majority of accidents are occured in CA</li>
    <li>All the weather terms</li>
    <li>Precipitation</li>
    <li>Side i.e. 'L' and 'R' columns</li>
    <li>Visibility(mi)</li>
    <li>Temperature(F)</li>
    <li>Wind_speed(mph)</li>
    <li>Junction</li>
</ul>

```python
dummy = pd.get_dummies(df["Side"])
dummy = dummy[['L', 'R']]
df = pd.concat([df, dummy], axis=1)
df.drop("Side", axis=1, inplace=True)
```

```python
variable = ["Distance(mi)","Temperature(F)", "Start_Lat" ,"Clear","Cloud", "Rain", "Heavy_Rain", "Heavy_Snow","Fog", "Month","Humidity(%)",
           "Pressure(in)","Visibility(mi)","Wind_Speed(mph)","Precipitation(in)","Traffic_Signal","Junction","Weekday", "Hour", "L", "R"]
```

```python
from sklearn import preprocessing
X = df[variable]
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
y = df['Severity'].values
y
```

<h3>Let's split our model into training data and testing data</h3>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
```

<p>Consider Random Forest Regressor Model</p>

```python
from sklearn.ensemble import RandomForestRegressor
rf_clf = RandomForestRegressor(max_depth=40,n_estimators=50)
rf_clf.fit(X_train,y_train)
```

```python
y_pred = list(map(round, rf_clf.predict(X_test)))
```

```python
print('Accuracy score of the Random Forest Regressor is: ', accuracy_score(y_pred, y_test))
```

```python
print('Jaccard score of Random Forest is: ', jaccard_score(y_true=y_test, y_pred=y_pred, average='weighted'))
print('F1 score of Random Forest is: ', f1_score(y_true=y_test, y_pred=y_pred, average='weighted'))
```

<p>Consider Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
tree1 = DecisionTreeClassifier(criterion='entropy', max_depth=None)
tree1.fit(X_train, y_train)
```

```python
tree_pred = tree1.predict(X_test)

print('Accuracy score of Decision tree model is: ', accuracy_score(tree_pred, y_test))
print('Jaccard score of Decision tree model is: ', jaccard_score(y_true=y_test, y_pred=tree_pred, average='weighted'))
print('F1 score of Decision tree model is: ', f1_score(y_true=y_test, y_pred=tree_pred, average='weighted'))
```

Consider Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss

LR3 = LogisticRegression(C = 0.001, solver = 'newton-cg', max_iter=1000)
LR3.fit(X_train, y_train)

y_pred1 = LR3.predict(X_test)
y_pred_proba = LR3.predict_proba(X_test)
jaccard_score1 = jaccard_score(y_test, y_pred1,average='weighted')
f1_score1 = f1_score(y_test, y_pred1, average='weighted')
```

```python
print('Accuracy score of Logistic Regression Model is: ', accuracy_score(y_pred=y_pred1, y_true=y_test))
```

```python
print('Jaccard score of Logistic Regression model is: ', jaccard_score(y_true=y_test, y_pred=y_pred1, average='weighted'))
print('F1 score of Logistic Regression model is: ', f1_score(y_true=y_test, y_pred=y_pred1, average='weighted'))
```
